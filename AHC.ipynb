```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
ncidata = np.genfromtxt('ncidata.txt')
print(ncidata.shape)
```

```python
ncidata_t = ncidata.T
print(ncidata_t)
print('-----')
print(ncidata_t.shape)
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Single")
dend = shc.dendrogram(shc.linkage(ncidata_t, method = 'single'))
```

```python
plt.figure(figsize=(10,7))
plt.title("Complete")
dend = shc.dendrogram(shc.linkage(ncidata_t, method = 'complete'))
```

```python
plt.figure(figsize=(10,7))
plt.title("Average")
dend = shc.dendrogram(shc.linkage(ncidata_t, method = 'average'))
```

```python
plt.figure(figsize=(10,7))
plt.title("Ward")
dend = shc.dendrogram(shc.linkage(ncidata_t, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

ahc_single = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="single")
ahc_single.fit_predict(ncidata_t)
```

```python
ahc_complete = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="complete")
ahc_complete.fit_predict(ncidata_t)
```

```python
ahc_average = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="average")
ahc_average.fit_predict(ncidata_t)
```

```python
ahc_ward = AgglomerativeClustering(n_clusters=5, affinity="euclidean", linkage="ward")
ahc_ward.fit_predict(ncidata_t)
```
