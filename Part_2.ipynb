```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DS - Part2 - Basketball.csv',header=0)
df.head(10)
```

```python
# Summary
df.info()
```

```python
df.shape
```

```python
# Replcaing '-' with '0'
df.replace('-',0,inplace=True)
```

```python
df.head(10)
```

```python
# Teams that started playing between 1930 and 1970
df['TeamLaunch'] = df['TeamLaunch'].astype(str)
df_year1929 = df[df['TeamLaunch'].str[:4].between('1930','1970')]
df_year1929[['Team','TeamLaunch']]
```

```python
# Teams that started playing in the year 1929
df[df.TeamLaunch == '1929']
```

```python
# Best performing team by Score
df['Score'] = df['Score'].astype(int)
df.sort_values(by='Score', ascending=False, inplace=False)
```

```python
# Best performing team by Baskets scored.
df['BasketScored'] = df['BasketScored'].astype(int)
df.sort_values(by='BasketScored', ascending=False, inplace=False)
```

* Team1 has the highest score and highest number of baskets scored across all the tournaments.

```python
# Guaging Winning percentage of the teams.

df['PlayedGames'] = df['PlayedGames'].astype(int)
df['WonGames'] = df['WonGames'].astype(int)

df['Win%'] = (df['WonGames']/df['PlayedGames'])* 100

df['Win%'].fillna(0,inplace = True)

df[['Team','Win%']]
```

* Team1 is again has the highest Win percentage of all the teams.

```python
# Highest position held by the teams.

df_bestteam = df[['Team','HighestPositionHeld']].groupby('HighestPositionHeld')
```
